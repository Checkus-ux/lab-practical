```python
students = {}

def add_student():

    ids = int(input("Enter ID Number: "))

    if ids in students:  
        print("Student ID already exists\n")  
        return  
        
    name = input("Enter Student Name: ")  
    age = int(input("Enter Age: "))  
    course = input("Enter Course: ")  
    
    students[ids] = {  
        "name": name,  
        "age": age,  
        "course": course
    }  
    print("Student added successfully!\n")

def view_student():
    if not students:
        print("No students available.\n")
        return
        
    print(" ALL STUDENTS ")
    for ids in students:
        print("ID:", ids)
        print("Name:", students[ids]['name'])
        print("Age:", students[ids]['age'])
        print("Course:", students[ids]['course'])
        print("-" * 20)
    print()

def search_student():
    if not students:
        print("No students available.\n")
        return
        
    id_s = int(input("Enter ID Number to search: "))
    if id_s in students:
        print("\n--- STUDENT FOUND ---")
        print("ID:", id_s)
        print("Name:", students[id_s]['name'])
        print("Age:", students[id_s]['age'])
        print("Course:", students[id_s]['course'], "\n")
    else:
        print("Student ID not found.\n")

def get_unique_course():
    if not students:
        print("No students available to extract courses from.\n")
        return
        
    courses = set()
    for rolls in students.values():
        courses.add(rolls['course'])
    print("Unique Courses:", courses, "\n")

def delete_student():
    if not students:
        print("No students available to delete.\n")
        return
        
    id_s = int(input("Enter ID Number to delete: "))
    if id_s in students:
        del students[id_s]
        print("Student deleted successfully!\n")
    else:
        print("Student not found.\n")


while True:
    print("=== Student Management System ===")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Get Unique Courses")
    print("5. Delete Student")
    print("6. Exit")

    try:
        choice = int(input("\nEnter your choice (1-6): "))
        print() 
        
        if choice == 1:  
            add_student()  
        elif choice == 2:  
            view_student()  
        elif choice == 3:  
            search_student()  
        elif choice == 4:  
            get_unique_course()  
        elif choice == 5:  
            delete_student()  
        elif choice == 6:  
            print("Exiting program. Goodbye!")
            break  
        else:  
            print("Invalid choice! Please enter a number between 1 and 6.\n")
            
    except ValueError:
        print("Invalid input! Please enter a valid number.\n")
```
