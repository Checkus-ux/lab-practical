```python

students = {}


def get_int_input(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print(" Invalid input. Please enter a valid number.")

def add_student():
    ids = get_int_input("Enter ID Number: ")
    

    if ids in students:
        print(" A student with this ID already exists!\n")
        return
        
    name = input("Enter Student Name: ").strip()
    age = get_int_input("Enter Age: ")
    course = input("Enter Course: ").strip()
    
    students[ids] = {
        "name": name,
        "age": age,
        "course": course
    }
    print(" Student added successfully!\n")

def view_student():
    if not students:
        print(" No students available.\n")
        return
        
    print("\n--- ALL STUDENTS ---")
    for ids in students:
        print(f"ID: {ids}")
        print(f"Name: {students[ids]['name']}")
        print(f"Age: {students[ids]['age']}")
        print(f"Course: {students[ids]['course']}")
        print("-" * 20)
    print()

def search_student():
    if not students:
        print(" No students available.\n")
        return
        
    id_s = get_int_input("Enter ID Number to search: ")
    
    if id_s in students:
        print("\n--- STUDENT FOUND ---")
        print(f"ID: {id_s}")
        print(f"Name: {students[id_s]['name']}")
        print(f"Age: {students[id_s]['age']}")
        print(f"Course: {students[id_s]['course']}\n")
    else:
        print(" Student not found.\n")

def get_unique_course():
    if not students:
        print("No students enrolled yet.\n")
        return
        
    # S
    courses = {student['course'] for student in students.values()}
    print(f" Unique Courses offered: {list(courses)}\n")

def delete_student():
    if not students:
        print(" No students available to delete.\n")
        return
        
    id_s = get_int_input("Enter ID Number to delete: ")
    
    if id_s in students:
        del students[id_s]
        print(" Student deleted successfully!\n")
    else:
        print(" Student not found.\n")


while True:
    print("=== Student Management System ===")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Get Unique Courses")
    print("5. Delete Student")
    print("6. Exit")

    choice = get_int_input("\nEnter your choice (1-6): ")  
    print() 
      
    if choice == 1:  
        add_student()  
    elif choice == 2:  
        view_student()  
    elif choice == 3:  
        search_student()  
    elif choice == 4:  
        get_unique_course()  
    elif choice == 5:  
        delete_student()  
    elif choice == 6:  
        print("Exiting program. Goodbye!")
        break  
    else:  
        print(" Invalid choice! Please select a number between 1 and 6.\n")
```
